```python
from mmdet.datasets import build_dataloader, build_dataset

import mmcv
from mmcv.runner import load_checkpoint
from mmdet.models import build_detector
from mmdet.apis import inference_detector, show_result
```

```python

cfg = mmcv.Config.fromfile('configs/pascal_voc/ssd300_voc.py')
cfg.model.pretrained = None

```

```python
# construct the model and load checkpoint
model = build_detector(cfg.model, test_cfg=cfg.test_cfg)
_ = load_checkpoint(model, 'work_dirs/ssd300_voc/epoch_48.pth')
```

```python
device = next(model.parameters()).device
```

```python
device
```

```python
model.cuda()
```

```python
dataset = build_dataset(cfg.data.test)
model.cfg = cfg
```

```python
imgs = [f"{dataset.img_prefix}JPEGImages/{x}.jpg" for x in dataset.img_ids]
results_paths = [x.replace('JPEGImages', 'pseudo') for x in imgs]
```

```python
for i, result in enumerate(inference_detector(model, imgs)):
    print(i, imgs[i])
    show_result(imgs[i], result, class_names=dataset.CLASSES, out_file=results_paths[i])
```
